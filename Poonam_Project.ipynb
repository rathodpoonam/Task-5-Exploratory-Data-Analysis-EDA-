```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
# 2. Load Dataset
df = pd.read_csv("train.csv")
```

```python
# 3. Basic Data Overview
print("First 5 rows of dataset:")
display(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nStatistical Summary:")
display(df.describe(include='all'))
```

```python
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

# # ---------------------------
# UNIVARIATE ANALYSIS
# ---------------------------

```python
# Survival count
plt.figure(figsize=(8,5))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.show()
# Observation: More passengers did not survive than survived.
```

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title("Age Distribution")
plt.show()
# Observation: Most passengers were between 20-40 years old.
```

```python
# Passenger class distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', data=df, palette='Set1')
plt.title("Passenger Class Distribution")
plt.show()
# Observation: Most passengers were in 3rd class.
```

```python
# Gender distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', data=df, palette='pastel')
plt.title("Gender Distribution")
plt.show()
# Observation: There were more male passengers than female passengers.
```

# # ---------------------------
# BIVARIATE ANALYSIS
# ---------------------------

```python
# Survival rate by gender
plt.figure(figsize=(8,5))
sns.barplot(x='Sex', y='Survived', data=df, palette='Set2')
plt.title("Survival Rate by Gender")
plt.show()
# Observation: Females had a much higher survival rate.
```

```python
# Survival rate by passenger class
plt.figure(figsize=(8,5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='Set1')
plt.title("Survival Rate by Passenger Class")
plt.show()
# Observation: Passengers in 1st class had the highest survival rate.
```

```python
# Age vs Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
plt.title("Age vs Survival")
plt.show()
# Observation: Younger passengers tended to survive more.
```

```python
# Fare distribution by survival
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Fare', hue='Survived', bins=30, kde=True)
plt.title("Fare Distribution by Survival")
plt.show()
# Observation: Higher fare was associated with higher survival probability.
```

# # ---------------------------
# MULTIVARIATE ANALYSIS
# ---------------------------


```python
# Correlation heatmap for numeric columns only
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
# Observation: Fare and Pclass are negatively correlated; Pclass and survival are strongly related.
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.show()
```

## ---------------------------
# HANDLING MISSING DATA
# ---------------------------

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

## ---------------------------
# FINAL SUMMARY OF FINDINGS
# ---------------------------


```python
summary = """
1. Females had a significantly higher survival rate compared to males.
2. 1st class passengers had the highest survival rate, followed by 2nd and 3rd class.
3. Younger passengers (especially children) had better survival chances.
4. Higher fare correlated positively with survival probability.
5. Most passengers were in 3rd class and were male.
6. Missing values were handled for 'Age' (median) and 'Embarked' (mode).
7. Pclass is strongly negatively correlated with Fare and positively associated with survival.
"""
print(summary)
```
